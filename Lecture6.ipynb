- Создайте класс Point(x, y) x, y - координаты точки на плоскости.
- Реализуйте два метода a.distanceTo(b) - евкливодов расстояние между $a$ и $b$
- Переопределите __str__ так чтобы при печати экземпляра выводилось (x, y) - то есть координаты точки

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def  adistanceTob(self,other):
        return str("S="+ str(((self.x-other.x)**2+(self.y-other.y)**2)**(1/2)))
    def __str__(self):
        return str("(x,y)=(" + str(self.x) + "," + str(self.y) +  ")")
        return str("(x,y)=(" + str(other.x) + "," + str(other.y) +  ")")
point1=Point(6,1)
point2=Point(3,0)
print(Point.adistanceTob(point1,point2))
print(Point.__str__(point1))
print(Point.__str__(point2))
```
